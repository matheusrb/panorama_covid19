# Panorama do COVID-19 no Brasil

COVID-19 é uma doença infecciosa causada por um novo tipo de coronavírus.

Transmitido principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, a gravidade dos sintomas varia muito de pessoa para pessoa.

Fato é, não se sabe muita coisa a respeito do COVID-19. Estudos estão sendo realizados no mundo todo, porém os resultados ainda não são conclusivos e definitivos.

<center><img alt="Colaboratory logo" width="50%" src="https://static.poder360.com.br/2022/12/coronavirus-imagem-1-848x477.jpg"></center>

Até o presente momento, observa-se que cerca de 80% dos casos confirmados são assintomáticos e rápidos. A maioria das pessoas que se encaixam nesse grupo, se recupera sem nenhuma sequela.

No entanto, 15% das pessoas terão infecções graves e precisarão de oxigênio. O restante das pessoas, que representam 5%, serão classificadas como infecções muito graves e precisarão de ventilação assistida, por meio de respiradores mecânicos em ambiente hospitalar.

Com o objetivo de elevar a consciência situacional a respeito do COVID-19 no Brasil, irei realizar uma análise sobre os dados públicos da doença.


Como complemento a análise, foi utilizada a base de dados referente ao preço do combustível e dados relacionados ao IGPM - Índice Geral de Preços do Mercado, para avlaiarmos os impactos econômicos peírodo a período.

# Dados Combustíveis

## Importação e filtragem da base de dados

```
df_combustivel = pd.DataFrame()

# importar o dataset COVID-19 para um DataaFrame
enderecos = ['https://raw.githubusercontent.com/matheusrb/panorama_covid19/master/Dados%20Combustível/ca-2019-02.csv',
             'https://raw.githubusercontent.com/matheusrb/panorama_covid19/master/Dados%20Combustível/ca-2020-01.csv',
             'https://raw.githubusercontent.com/matheusrb/panorama_covid19/master/Dados%20Combustível/ca-2020-02.csv',
             'https://raw.githubusercontent.com/matheusrb/panorama_covid19/master/Dados%20Combustível/ca-2021-01.csv',
             'https://raw.githubusercontent.com/matheusrb/panorama_covid19/master/Dados%20Combustível/ca-2021-02.csv',
             'https://raw.githubusercontent.com/matheusrb/panorama_covid19/master/Dados%20Combustível/ca-2022-01.csv',
             'https://raw.githubusercontent.com/matheusrb/panorama_covid19/master/Dados%20Combustível/ca-2022-02.csv',
             'https://raw.githubusercontent.com/matheusrb/panorama_covid19/master/Dados%20Combustível/ca-2023-01.csv'
            ]

for i, endereco in enumerate(enderecos):
  # Baixando o conteúdo do arquivo
  response = requests.get(enderecos[i])
  content = response.content

  # Detectando a codificação do conteúdo usando chardet
  result = chardet.detect(content)
  encoding = result['encoding']

  # Decodificando o conteúdo para UTF-8
  content_utf8 = content.decode(encoding)

  # Lendo o conteúdo decodificado com o Pandas
  df_combustivel_temp = pd.read_csv(StringIO(content_utf8), sep=";")
  df_combustivel = pd.concat([df_combustivel, df_combustivel_temp])

df_combustivel['Data da Coleta'] = pd.to_datetime(df_combustivel['Data da Coleta'], format='%d/%m/%Y')
df_combustivel.sort_values(by='Data da Coleta', inplace=True)
df_combustivel.head()
```

```
#Elimina colunas irrelevantes para a análise
df_combustivel = df_combustivel[['Data da Coleta', 'Produto', 'Valor de Venda']]
df_combustivel.head()
```

```
#corrige o data type do valor de venda
df_combustivel['Valor de Venda'] = df_combustivel['Valor de Venda'].str.replace(',', '.').astype(float)
df_combustivel.dtypes
```

```
#Organiza o data frame por ordem cronológica
df_combustivel.reset_index(drop=True, inplace=True)
df_combustivel.sort_values(by='Data da Coleta')
df_combustivel.head()
```

## Criação do preço médio por data e tipo de produto

```
df_combustivel_medio = df_combustivel.groupby(['Data da Coleta', 'Produto'], as_index=False).agg({'Valor de Venda' : 'mean'})
df_combustivel_medio.reset_index(drop=True, inplace=True)
df_combustivel_medio.sort_values(by='Data da Coleta')
df_combustivel_medio.head(20)
```

### Criação das Médias Móveis dos Preços dos Combustíveis

#### Diesel

```
#Criação da media movel de 14 dias diesel
df_diesel_medio = df_combustivel_medio.loc[df_combustivel_medio['Produto'] == 'DIESEL']
df_diesel_medio.reset_index(drop=True, inplace=True)
df_diesel_medio['media_movel'] = df_diesel_medio['Valor de Venda'].rolling(14).mean()
df_diesel_medio.media_movel = df_diesel_medio.media_movel.fillna(df_diesel_medio.loc[13,'media_movel'])

df_diesel_medio.head(20)
```

```
#Criação da media movel de 14 dias diesel-s10
df_diesel_s10_medio = df_combustivel_medio.loc[df_combustivel_medio['Produto'] == 'DIESEL S10']
df_diesel_s10_medio.reset_index(drop=True, inplace=True)
df_diesel_s10_medio['media_movel'] = df_diesel_s10_medio['Valor de Venda'].rolling(14).mean()
df_diesel_s10_medio.media_movel = df_diesel_s10_medio.media_movel.fillna(df_diesel_s10_medio.loc[13,'media_movel'])

df_diesel_s10_medio.head()
```

```
#Criação da media movel de 14 dias gasolina
df_gasolina_medio = df_combustivel_medio.loc[df_combustivel_medio['Produto'] == 'GASOLINA']
df_gasolina_medio.reset_index(drop=True, inplace=True)
df_gasolina_medio['media_movel'] = df_gasolina_medio['Valor de Venda'].rolling(14).mean()
df_gasolina_medio.media_movel = df_gasolina_medio.media_movel.fillna(df_gasolina_medio.loc[13,'media_movel'])

df_gasolina_medio.head()
```

Inicialmente foi necessário tratar valores de novos casos para que valores nulos fossem preenchidos pelo seu antecessor, algo possível devido ao bom volume de dados de qualidade existentes.

### Criação médias móveis - Brasil
Criação de médias moveis de novos casos e novos óbitos no Brasil

```
#Tratamento para criação da media movel
df_brasil.new_cases = df_brasil.new_cases.fillna(0)
df_brasil.new_deaths = df_brasil.new_deaths.fillna(0)

#definição do período da média móvel
dias_media_movel = 14

#Criação da media movel de 14 dias
df_brasil['media_movel_casos'] = df_brasil.new_cases.rolling(dias_media_movel).mean()
df_brasil.media_movel_casos = df_brasil.media_movel_casos.fillna(0)

#Criação da media movel de 14 dias
df_brasil['media_movel_obitos'] = df_brasil.new_deaths.rolling(dias_media_movel).mean()
df_brasil.media_movel_obitos = df_brasil.media_movel_obitos.fillna(0)
df_brasil.head(100)
```

```
import calendar
# Usar a função melt para transformar a tabela
df_transformado = pd.melt(df_igpm, id_vars=['Unnamed: 0'], value_vars=['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez'],
                           var_name='mes', value_name='valor')

# Juntar o mês e o ano em uma única coluna
df_transformado['mes_ano'] = df_transformado['mes'] + ' ' + df_transformado['Unnamed: 0'].astype(str)

# Excluir as colunas 'mes' e 'Unnamed: 0'
df_transformado = df_transformado.drop(['mes', 'Unnamed: 0'], axis=1)

# Dicionário de tradução de nomes de meses em português para inglês
meses_dict = {
    'Jan': 'January',
    'Fev': 'February',
    'Mar': 'March',
    'Abr': 'April',
    'Mai': 'May',
    'Jun': 'June',
    'Jul': 'July',
    'Ago': 'August',
    'Set': 'September',
    'Out': 'October',
    'Nov': 'November',
    'Dez': 'December'
}

# Aplicar a tradução dos nomes de meses
df_transformado['mes_ano'] = df_transformado['mes_ano'].apply(lambda x: ' '.join([meses_dict[mes] if mes in meses_dict else mes for mes in x.split()]))

# Transformar a coluna 'mes_ano' em um objeto datetime com o último dia de cada mês
df_transformado['mes_ano'] = pd.to_datetime(df_transformado['mes_ano'], format='%B %Y') + pd.offsets.MonthEnd()

df_transformado = df_transformado.sort_values(by='mes_ano')

# Criar uma nova coluna 'acumulado' com os valores acumulados da coluna 'valor'
df_transformado['acumulado_igpm'] = df_transformado['valor'].cumsum()

df_transformado.head()
```

## Conclusão

É indiscutível o quanto a pandemia impactou a maneira como se vive no mundo, bem como suas sequelas ainda serão vistas por mais algum tempo, especialmente em termos sociais, uma vez que houve alta na taxa de desempregados, defasagem no ensino e impactos econômicos como inflação e, por consequencia, uma alta das taxas de juros e redução do poder de comsumo.
Utilizando os dados é possível inferir que, apesar do atraso em relação ao mundo no que se diz respeito a reação, o Brasil ainda teve menos óbitos relativos que seus vizinhos.
Também é interessante notar que há uma relação entre o aumento do preço dos combustíveis e o aumento do número de casos. Com a insegurança socio-política causada pela COVID é natural o movimento de saída de investimentos de países com moedas mais fracas e sua realocação em moeda mais forte, gerando aumento na cotação do Dólar, bem como desvalorização do Real. Tal aumento acarretou no aumento dos preços de combustíveis, uma vez que na época, a Petrobrás - maior beneficiadora de petroleo do país - ainda tinha vigente sua politica de paridade de preço internacional.

Em termos sanitários observa-se que, mesmo com as vacinações tendo começado no começo de 2021,  ainda assim não foi suficiente para conter a nova variante Omicron, que elevou o número de mortos de maneira substancialmente agressiva logo no início de 2022. De maneira expeculativa é possível questionar se esta alta de casos causada pela Omicron também está associada a um relaxamento no que se diz respeito aos cuidados exigidos ao longo da pandemia, bem como às festas de fim de ano como potencializadores do número de casos e, por consequencia, óbitos no mundo todo registrado nesta época.

Por fim, é nítido que a crise sanitária vem sendo controlada no Brasil, tal evidencia se dá quando comparadas as variações entre as médias móveis do Brasil e do mundo no pico mais recente de COVID, em que é possível notar uma variação muito mais suave em terras tupiniquins que no mundo como um todo.

Apesar de o controle da crise e das sequelas citadas no inicio do texto, ficaram alguns hábitos benéficos. De maneira geral, a pandemia deixou uma lição de higiêne e cuidado pessoal, bem como trouxe novas culturas como a de se utilizar máscara para se proteger e proteger o outro, algo que perdura até então.
