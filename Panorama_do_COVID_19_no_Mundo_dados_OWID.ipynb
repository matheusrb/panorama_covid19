# Panorama do COVID-19 no Brasil

COVID-19 é uma doença infecciosa causada por recém-descoberto coronavírus.

Transmitido principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, a gravidade dos sintomas varia muito de pessoa para pessoa.

Fato é, não se sabe muita coisa a respeito do COVID-19. Estudos estão sendo realizados no mundo todo, porém os resultados ainda não são conclusivos e definitivos.

<center><img alt="Colaboratory logo" width="50%" src="https://static.poder360.com.br/2022/12/coronavirus-imagem-1-848x477.jpg"></center>

Até o presente momento, observa-se que cerca de 80% dos casos confirmados são assintomáticos e rápidos. A maioria das pessoas que se encaixam nesse grupo, se recupera sem nenhuma sequela.

No entanto, 15% das pessoas terão infecções graves e precisarão de oxigênio. O restante das pessoas, que representam 5%, serão classificadas como infecções muito graves e precisarão de ventilação assistida, por meio de respiradores mecânicos em ambiente hospitalar.

Com o objetivo de elevar a consciência situacional a respeito do COVID-19 no Brasil, irei realizar uma análise sobre os dados públicos da doença.

As colunas que possuem mais de 25% de valores nulos sao eliminadas para facilitar a manipulação e o processamento dos dados, uma vez que devido a carencia de dados, trazem pouqíssima relevancia à análise.

```
df_sa = df.loc[df.continent == 'South America'].sort_values(by='date', ascending=False)
```

```
df_paises_sa = df_sa.groupby(['location'], as_index=False).agg({'total_cases' : 'max',
                                                                'total_deaths' : 'max',
                                                                'gdp_per_capita' : 'first',
                                                                'human_development_index' : 'first',
                                                                'population' : 'first'}).sort_values(by='total_deaths')
df_paises_sa['per_rel_cases'] = df_paises_sa.total_cases/df_paises_sa.population*100
df_paises_sa['per_rel_deaths'] = df_paises_sa.total_deaths/df_paises_sa.population*100
df_paises_sa.dropna(inplace=True)
```

## Evolução dos casos
Abaixo segue um gráfico da evolução diária de novos casos, foi usada também uma média móvel que filtra os valores e torna possível a visualização de uma tendência mais clara, especialmente quando são utilizadas janelas de tempo menores.

Inicialmente foi necessário tratar valores de novos casos para que valores nulos fossem preenchidos pelo seu antecessor, algo possível devido ao bom volume de dados de qualidade existentes.

```
#Tratamento para criação da media movel
df_brasil.new_cases = df_brasil.new_cases.fillna(method='bfill')

#definição do período da média móvel
dias_media_movel = 14

#Criação da media movel de 7 dias
df_brasil['media_movel'] = df_brasil.new_cases.rolling(dias_media_movel).mean()
df_brasil.media_movel = df_brasil.media_movel.fillna(0)
df_brasil.head()
```

```
#Cria uma média movel do mundo todo para fins de comparação em termos de tendencia
df_world = df.groupby(['date'], as_index=False).agg({'new_cases' : 'sum'})
#Tratamento para criação da media movel
df_world.new_cases = df_world.new_cases.fillna(method='bfill')

#definição do período da média móvel
dias_media_movel = 14

#Criação da media movel de 7 dias
df_world['media_movel'] = df_world.new_cases.rolling(dias_media_movel).mean()
df_world.media_movel = df_world.media_movel.fillna(0)
df_world.head()
```

```
df_brasil.sort_values(by='date', inplace=True)
fig, ax1 = plt.subplots(figsize=(15, 8))

#plot novos obitos diário
ax1.plot(df_brasil['date'], df_brasil['new_cases'], label='Novos óbitos Brasil')
#plot MM de obitos
ax1.plot(df_brasil['date'], df_brasil['media_movel'], label='Média móvel de 7 dias de novos óbitos Brasil', color='r')
ax1.grid(visible=True)
ax1.set_xlabel('Data')
ax1.set_ylabel('Novos Óbitos')
#ax1.set_title('Tendencia de óbitos no Brasil + ')

#Criação da curva do mundo
ax2 = ax1.twinx()

#plot novos obitos diário
#ax2.plot(df_world['date'], df_world['new_cases'], label='Novos óbitos mundo', color='g')
#plot MM de obitos
ax2.plot(df_world['date'], df_world['media_movel'], label='Média móvel de 2 semanas de novos óbitos mundo', color='purple')
ax2.grid(visible=True, linestyle="dashed")
ax2.set_xlabel('Data')
ax2.set_ylabel('Novos Óbitos no Mundo')

# Definindo a janela de tempo a ser destacada
start_time = '2021-02-08'
end_time = '2021-02-16'

# Adicionando um retângulo para destacar a janela de tempo
ax2.axvspan(start_time, end_time, color='gray', alpha=0.3, label='Pré e Carnaval 2021')

# Definindo a janela de tempo a ser destacada
start_time = '2021-02-27'
end_time = '2021-03-07'

# Adicionando um retângulo para destacar a janela de tempo
ax2.axvspan(start_time, end_time, color='red', alpha=0.2, label='Média de tempo infecção ao óbito (19 dias)')


# Definindo a janela de tempo a ser destacada
start_time = '2021-01-19'
end_time = '2021-01-20'

# Adicionando um retângulo para destacar a janela de tempo
ax2.axvspan(start_time, end_time, color='green', alpha=1, label='Início da vacinação')



# Definindo a janela de tempo a ser destacada
start_time = '2021-12-03'
end_time = '2021-12-04'

# Adicionando um retângulo para destacar a janela de tempo
ax2.axvspan(start_time, end_time, color='red', alpha=1, label='Primeiro caso Omicron')

#seta legenda
curvas, labels = ax1.get_legend_handles_labels()
curva2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(curvas + curva2, labels + labels2, loc='upper left')

plt.tight_layout()
plt.show()
```

## Conclusão

É indiscutível o quanto a pandemia impactou a maneira como se vive no mundo, bem como suas sequelas ainda serão vistas por mais algum tempo, especialmente em termos sociais, uma vez que houve alta na taxa de desempregados, defasagem no ensino e impactos econômicos como inflação e, por consequencia, uma alta das taxas de juros e redução do poder de comsumo.
Utilizando os dados é possível inferir que, apesar do atraso em relação ao mundo no que se diz respeito a reação, o Brasil ainda teve menos óbitos relativos que seus vizinhos.
É interessante notar também que houve um crescente acentuado de óbitos após a semana de Pré e Carnaval, considerando o tempo da infecção até o óbito como 19 dias, segundo [este artigo](https://ciis.fmrp.usp.br/covid19/estimativa-de-populacao-infectada/#:~:text=O%20tempo%20médio%20da%2C%20infecção,para%20evoluir%20para%20o%20óbito.) produzido pela USP.  
Por outro lado, mesmo com as vacinações tendo começado no começo de 2021, não foi suficiente para conter a nova variante Omicron, que elevou o número de mortos de maneira muito agressiva já no início de 2022. De maneira expeculativa é possível questionar se esta alta de casos causada pela Omicron também está associada a um relaxamento no que se diz respeito aos cuidados exigidos ao longo da pandemia, bem como às festas de fim de ano como potencializadores do número de óbitos no mundo todo registrado nesta época.
Por fim, é nítido que a crise sanitária vem sendo controlada no Brasil, tal evidencia se dá quando comparadas as variações entre as médias móveis do Brasil e do mundo no pico mais recente de COVID, em que é possível notar uma variação muito mais sauve em terras tupiniquins que no mundo como um todo.
Apesar de o controle da crise e das sequelas citadas no inicio do texto, ficaram alguns habitos benéficos. De maneira geral, a pandemia deixou uma lição de higiêne e cuidado pessoal, bem como trouxe novas culturas como a de se utilizar máscara para se proteger e proteger o outro, algo que perdura até então.
